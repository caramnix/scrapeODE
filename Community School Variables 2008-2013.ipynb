```python
def get_school_list(year, final_val): 
    time.sleep(1)
    driver = webdriver.Chrome(executable_path="C:\chromedriver.exe")
    driver.implicitly_wait(0.5)
    url = "https://reports.education.ohio.gov/finance/legacy-report" 
    driver.get(url)
    driver.implicitly_wait(10)
    time.sleep(3)

    ## scrape 2000
    YEAR = year
    FINAL = final_val
    #first step, fill in info, press go 

    # fill in LEA type 
    text_box_path = '/html/body/app-root/eas-ui-lib-spinner/div/eas-core-application-container/div/div[2]/as-split/as-split-area[2]/div/div/app-legacy-reports/lib-legacy-report/div/div/div/div[2]/ul/li[1]/eas-ui-lib-form-group/label/ng-select/div/div/div[3]/input'
    text_box = driver.find_element(By.XPATH, text_box_path)
    text_box.send_keys("Traditional School District")

    text_box_selection_path = "/html/body/app-root/eas-ui-lib-spinner/div/eas-core-application-container/div/div[2]/as-split/as-split-area[2]/div/div/app-legacy-reports/lib-legacy-report/div/div/div/div[2]/ul/li[1]/eas-ui-lib-form-group/label/ng-select/ng-dropdown-panel/div/div[2]/div"
    time.sleep(.25)
    drop_box_1 = driver.find_element(By.XPATH, text_box_selection_path)
    time.sleep(.25)
    drop_box_1.click()

    #fill in year 

    year_text_box_path ="/html/body/app-root/eas-ui-lib-spinner/div/eas-core-application-container/div/div[2]/as-split/as-split-area[2]/div/div/app-legacy-reports/lib-legacy-report/div/div/div/div[2]/ul/li[2]/eas-ui-lib-form-group/label/ng-select/div/div/div[3]/input"
    year_text_box = driver.find_element(By.XPATH, year_text_box_path)
    year_text_box.send_keys(YEAR)

    year_drop_select ="/html/body/app-root/eas-ui-lib-spinner/div/eas-core-application-container/div/div[2]/as-split/as-split-area[2]/div/div/app-legacy-reports/lib-legacy-report/div[1]/div/div/div[2]/ul/li[2]/eas-ui-lib-form-group/label/ng-select/ng-dropdown-panel/div/div[2]/div/span"
    time.sleep(.25)
    drop_box_2 = driver.find_element(By.XPATH, year_drop_select)
    time.sleep(.25)
    drop_box_2.click()

    time.sleep(.25)
    go_path = "/html/body/app-root/eas-ui-lib-spinner/div/eas-core-application-container/div/div[2]/as-split/as-split-area[2]/div/div/app-legacy-reports/lib-legacy-report/div/div/div/div[2]/ul/li[3]/button"
    time.sleep(.25)
    go_button = driver.find_element(By.XPATH, go_path)
    time.sleep(.25)

    go_button_bool = None
    while not go_button_bool:
        try:
            go_button.click()
            go_button_bool=1
        except ElementClickInterceptedException:
            time.sleep(2)

    time.sleep(.25)
    #next step 

    payment_path = "/html/body/app-root/eas-ui-lib-spinner/div/eas-core-application-container/div/div[2]/as-split/as-split-area[2]/div/div/app-legacy-reports/lib-legacy-report/div[1]/div/div/div[3]/div/ul/li/eas-ui-lib-form-group/label/ng-select/div/div/div[3]/input"
    payment_path_box = driver.find_element(By.XPATH, payment_path)
    time.sleep(.25)

    payment_path_box.send_keys(FINAL)
    payment_path_selection_path = "/html/body/app-root/eas-ui-lib-spinner/div/eas-core-application-container/div/div[2]/as-split/as-split-area[2]/div/div/app-legacy-reports/lib-legacy-report/div[1]/div/div/div[3]/div/ul/li/eas-ui-lib-form-group/label/ng-select/ng-dropdown-panel/div/div[2]/div/span"

    time.sleep(.25)
    drop_box_3 = driver.find_element(By.XPATH, payment_path_selection_path)
    drop_box_3.click()

    #next step is to find all LEA values 

    lea_box_path = "/html/body/app-root/eas-ui-lib-spinner/div/eas-core-application-container/div/div[2]/as-split/as-split-area[2]/div/div/app-legacy-reports/lib-legacy-report/div[2]/div/div/div/div/div[1]/eas-ui-lib-form-group/label/ng-select/div/div/div[3]/input"
    lea_arrow_path = "/html/body/app-root/eas-ui-lib-spinner/div/eas-core-application-container/div/div[2]/as-split/as-split-area[2]/div/div/app-legacy-reports/lib-legacy-report/div[2]/div/div/div/div/div[1]/eas-ui-lib-form-group/label/ng-select/div"

    #click on arrow to show options 
    driver.find_element(By.XPATH, lea_arrow_path).click()

    time.sleep(.25)
    drop_down_path = "/html/body/app-root/eas-ui-lib-spinner/div/eas-core-application-container/div/div[2]/as-split/as-split-area[2]/div/div/app-legacy-reports/lib-legacy-report/div[2]/div/div/div/div/div[1]/eas-ui-lib-form-group/label/ng-select/ng-dropdown-panel/div"
    drop_down= driver.find_element(By.XPATH, drop_down_path)

    ddt= drop_down.text
    list_school_names_i = ddt.splitlines()
    
    return(list_school_names_i)
```
